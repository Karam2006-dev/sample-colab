```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
text="I love nlp."
word=word_tokenize(text)
print(word)
senten = sent_tokenize(text)
print(senten)
paragraph="""A paragraph is a group of sentences focusing on a single, unified idea, acting as a structural unit of writing to organize thoughts. Generally composed of a topic sentence, supporting details, and a concluding sentence, paragraphs enhance readability and structure. They range from one sentence in journalism to several in academic essays."""
sentenc=sent_tokenize(paragraph)
print(sentenc)
word=word_tokenize(paragraph)
print(word)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
words=["program","programs","programer","programing","programers"]
stem=[ps.stem(word) for word in words]
print(words)
print(stem)
```
